```python
df3=pd.merge(df2,df1,how='left',left_on=df2.object_id,right_on=df1.user_id)
df3.head()
```

```python
df3['date']=df3['time_stamp'].astype('datetime64')
df3['week']=df3['date'].dt.isocalendar().week
```

```python
grp_week=df3.groupby(['user_id','week']).visited.count().reset_index()
grp_week
```

```python
adopted_user = grp_week.loc[grp_week['visited'] >=3]
grp_week['adopted_user'] = grp_week['visited'].apply(lambda x: 1 if x >= 3 else 0)
```

```python
q = df3.groupby(['creation_source']).size().values
p = df3.groupby(['creation_source']).size().index
y = np.array(q)
x = np.array(p)
plt.pie(y, labels = x,explode= (0,0.1,0,0,0), autopct='%1.2f%%',shadow=True)
plt.title("Percentage distribution of types of creation source", color = 'm',fontsize=16)
plt.show()
```
